###  Display the least squares line on the scatterplot obtained in (d). Draw the population regression line on the plot, in a different color. Use the legend() method of the axes to create an appro- priate legend.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Use the updated epsilon
np.random.seed(0)
x = np.random.normal(0, 1, 100)
eps = np.random.normal(loc=0, scale=0.25, size=100)
y = -1 + 0.5 * x + eps

# Fit least squares linear model
x_reshaped = x.reshape(-1, 1)
model = LinearRegression().fit(x_reshaped, y)
y_hat = model.predict(x_reshaped)

# Population regression line (true relationship, no noise)
y_population = -1 + 0.5 * x

# Plotting
plt.figure(figsize=(8, 5))
sns.scatterplot(x=x, y=y, label="Data")
plt.plot(x, y_hat, color='red', label="Least Squares Fit")
plt.plot(x, y_population, color='green', linestyle='--', label="Population Regression Line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot with Least Squares and Population Regression Lines')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Assume x and y are already defined as in your previous steps
np.random.seed(0)
x = np.random.normal(0, 1, 100)
eps = np.random.normal(loc=0, scale=0.25, size=100)
y = -1 + 0.5 * x + eps

# Generate polynomial features: [1, x, x^2]
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(x.reshape(-1, 1))

# Fit polynomial regression model
model_poly = LinearRegression().fit(X_poly, y)
y_pred_poly = model_poly.predict(X_poly)

# Fit linear regression model for comparison
model_linear = LinearRegression().fit(x.reshape(-1, 1), y)
y_pred_linear = model_linear.predict(x.reshape(-1, 1))

# R-squared values for each model
r2_linear = r2_score(y, y_pred_linear)
r2_poly = r2_score(y, y_pred_poly)

print(f"Linear Model R²: {r2_linear:.4f}")
print(f"Polynomial Model R²: {r2_poly:.4f}")
print("Polynomial coefficients:", model_poly.coef_)
print("Polynomial intercept:", model_poly.intercept_)
```
