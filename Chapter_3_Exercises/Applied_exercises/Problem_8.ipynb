```python
import statsmodels.api as sm

X = df['horsepower']
X = sm.add_constant(X)
y = df['mpg']

model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```

```python
import pandas as pd
import statsmodels.api as sm

# Assume your DataFrame is called df
# Clean horsepower column (convert to numeric, drop missing)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df = df.dropna(subset=['horsepower', 'mpg'])

# Prepare X (with constant) and y
X = sm.add_constant(df['horsepower'])  # Adds intercept
y = df['mpg']

# Fit the model
model = sm.OLS(y, X)
results = model.fit()

# Prepare input for prediction (horsepower = 98)
new_X = pd.DataFrame({'const': 1, 'horsepower': [98]})

# Make prediction with intervals
prediction = results.get_prediction(new_X)
prediction_summary = prediction.summary_frame(alpha=0.05)  # 95% interval

print(prediction_summary)
```

### Plot the response and the predictor in a new set of axes ax. Use the ax.axline() method or the abline() function defined in the lab to display the least squares regression line.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# ASSUMING your DataFrame is 'df' and you have cleaned it as follows:
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df = df.dropna(subset=['horsepower', 'mpg'])

# Set up response and predictor
X = df['horsepower']
y = df['mpg']

# Fit the model
X_const = sm.add_constant(X)
model = sm.OLS(y, X_const)
results = model.fit()

# Get coefficients for the regression line
intercept = results.params['const']
slope = results.params['horsepower']

# PLOT
fig, ax = plt.subplots()
ax.scatter(X,y, color = 'blue', label = 'Data points')

ax.axline((0, intercept), slope=slope, color='red', label='Regression line')


ax.set_xlabel('Horsepower')
ax.set_ylabel('MPG')
ax.set_title('MPG vs Horsepower with Regression Line')
ax.legend()
plt.show()
```
