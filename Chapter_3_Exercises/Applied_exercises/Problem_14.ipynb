### What is the correlation between x1 and x2? Create a scatterplot displaying the relationship between the variables.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Data generation
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10

# Calculate correlation
correlation = np.corrcoef(x1, x2)[0, 1]
print("Correlation between x1 and x2:", correlation)

# Create scatterplot
df  = pd.DataFrame({"x1":x1,"x2":x2})
sns.scatterplot(data = df, x ="x1", y = "x2")
plt.title(f'Scatterplot of x1 vs x2 (Correlation:{correlation:.3f})')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show();
```

### (c) Using this data, fit a least squares regression to predict y using x1 and x2. Describe the results obtained. What are βˆ0, βˆ1, and βˆ2? How do these relate to the true β0, β1, and β2? Can you reject the null hypothesis H0 : β1 = 0? How about the null hypothesis H0 : β2 = 0?

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Data generation
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)

# Prepare predictors (add constant for intercept)
X = np.column_stack([x1, x2])
X = sm.add_constant(X)  # adds the intercept term

# Fit least squares regression
model = sm.OLS(y, X).fit()

# Get results
print(model.summary())

# To access estimated coefficients
beta0_hat = model.params[0]
beta1_hat = model.params[1]
beta2_hat = model.params[2]
print(f"Estimated β0: {beta0_hat:.2f}")
print(f"Estimated β1: {beta1_hat:.2f}")
print(f"Estimated β2: {beta2_hat:.2f}")
```

```python
# you would reject the null for the x1 but not for x2
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Add a new 'high leverage' observation to the original data
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)

# Add a new outlier / high leverage point
x1_new = np.append(x1, 0.1)
x2_new = np.append(x2, 0.8)
y_new  = np.append(y, 6)

df_new = pd.DataFrame({'x1': x1_new, 'x2': x2_new, 'y': y_new})

# Model (c): Predict y using x1 and x2
model_c = smf.ols('y ~ x1 + x2', data=df_new).fit()
print(model_c.summary())

# Model (d): Predict y using x1 only
model_d = smf.ols('y ~ x1', data=df_new).fit()
print(model_d.summary())

# Model (e): Predict y using x2 only
model_e = smf.ols('y ~ x2', data=df_new).fit()
print(model_e.summary())

# Identify if the new point is an outlier or high leverage (optional: influence plot)
import matplotlib.pyplot as plt
sm.graphics.influence_plot(model_c, criterion="cooks")
plt.show()
```
