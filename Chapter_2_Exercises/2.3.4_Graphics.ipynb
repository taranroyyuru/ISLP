```python
from matplotlib.pyplot import subplots
import numpy as np
rng = np.random.default_rng()
```

```python
fig,ax = subplots(figsize=(8,8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x,y)

#Use plots to create figures
# subplot returns a tuple of relevant axis objects
#plot is when we actually plot the method
```

```python
output = subplots(figsize =(8,8))
fig = output[0]
ax = output[1]
```

```python
# scatter plot
fig, ax = subplots(figsize = (8,8))
ax.plot(x,y,'o');
```

```python
# this is the better way to use it
fig,ax = subplots(figsize = (8,8))
ax.scatter(x,y, marker ='*');
#adding the semi-colon to the end otains the object d
```

```python
fig,ax = subplots(figsize = (8,8))
ax.scatter(x,y, marker ='*');
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-axis")
ax.set_title("Plots of X vs Y");
```

```python
#contour plot
#takes in A vecotr of x-values
# a vector of y-values
# A matrix whose elements correspond to the z value( the third dimension) for each pair of 
# (x,y) coordinates
```

```python
fig, ax = subplots(figsize = (8,8))
x = np.linspace(-np.pi,np.pi,50) # returns a vector of n numbers
y = x
f = np.multiply.outer(np.cos(y), 1/(1+x**2))
ax.contour(x,y,f);
```

```python
# can increase the resolution by adding more levels to the image
fig, ax = subplots(figsize = (8,8))
ax.contour(x,y,f,levels = 45);
```

```python
fig, ax = subplots(figsize = (8,8))
ax.imshow(f);

#it produces a color-coded plot whose colors depend on the z-value
#heatmap and used to plot temperatures in weather forecasting
```
