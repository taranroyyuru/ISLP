```python
idx_80 = Auto_re['year']>80
Auto_re.loc[idx_80,['weight','origin']]
```

```python
#lambda function
Auto_re.loc[lambda df: df['year'] >80, ['weight','origin']]
```

```python
Auto_re.loc[lambda df:(df['year']>80) &(df['mpg'] >30),
            ['weight','origin']
]

#Suppose that we want all cars built after 1980 that achieve greater than 30 miles per gallon
```

```python
# Want to retrieve all Ford and Datsun cars with dsiplacement less than 300
# We check whether each name entry containts either string ford or datsun using str.contatins() method

Auto_re.loc[lambda df: (df['displacement'] <300)
& (df.index.str.contains('ford')
   | df.index.str.contains('datsun')),
['weight','origin']
]
```
