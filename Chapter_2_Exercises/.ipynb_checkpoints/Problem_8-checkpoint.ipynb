```python
import matplotlib.pyplot as plt

# side‐by‐side boxplots of Outstate tuition for Private = Yes vs No
college.boxplot( column='Outstate', by='Private',figsize=(6, 4),grid=False)

plt.title("Out-of-State Tuition by Private/Public")  
plt.xlabel("Private School?")    
plt.ylabel("Out-of-State Tuition")
plt.tight_layout()
plt.show();
```

```python
df['Elite'] = pd.cut(df['Top10perc'],
                     [0,50,100],
                     labels=["No","Yes"])
```

```python
# pick four columns and a bin‐count for each
plots = [
    ('Apps',       10),   # # of applications, 10 bins
    ('Accept',     30),   # # of acceptances, 30 bins
    ('Outstate',   20),   # out-of-state tuition, 20 bins
    ('Enroll',     15),   # # who enrolled, 15 bins
]


fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()
for ax, (col, nbins) in zip(axes, plots):
    college[col].plot.hist(
        bins=nbins,
        ax=ax,
        alpha=0.7,
        edgecolor='black'
    )
    ax.set_title(f"{col} (bins={nbins})")
    ax.set_xlabel(col)
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
